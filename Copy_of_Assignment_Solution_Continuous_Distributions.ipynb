```
lower_limit = 50
upper_limit = 150
range = upper_limit - lower_limit
```

```
# P(Y > 74) = 1 - P(Y < 74)

1 - uniform.cdf(74, lower_limit, range)
```

```
# P(50 < Y < 126) = P(Y < 126) - P(Y < 50)

uniform.cdf(126, lower_limit, range) - uniform.cdf(50, lower_limit, range)
```

# Problem 5

Screenshot 2020-11-25 at 7.09.30 PM.png

```
# X ~ N(25, 36) 
mean = 25
standard_deviation = math.sqrt(36)

# (i) P(X < 28)

norm.cdf(28, mean, standard_deviation)
```

```
# (ii) P(X > 30) = 1 - P(X < 30)

1 - norm.cdf(30, mean, standard_deviation)
```

```
# (iii) P(X < 20)

norm.cdf(20, mean, standard_deviation)
```

```
# |X - 25| < 4 
# => (X - 25) < 4 or -(X - 25) < 4
# => X < 29 or X > 21

# (iv) P(|X - 25| < 4) = P(21 < X < 29) = P(X < 29) - P(X < 21)

norm.cdf(29, mean, standard_deviation) - norm.cdf(21, mean, standard_deviation)
```
